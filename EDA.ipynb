```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

## Uni variate Analysis

```python
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_setosa
```

```python
 df_virginica=df.loc[df['species']=='virginica']
```

```python
df_virginica
```

```python
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
df_versicolor
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']), '_')
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']), 'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']), 'x')
```

```python
plt.xlabel('Sepal Length')
plt.show()
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue="species",size=3)
```
